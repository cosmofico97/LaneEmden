$$
\frac{1}{\xi^2}\frac{d}{d\xi}\left(\xi^2 \, \frac{d\theta}{d\xi}\right) = -\theta^n\,, \quad 
\mathrm{with} \;\; \theta(0)=1, \; \frac{d\theta}{d\xi}(0)=0
$$

$$
\begin{cases}
\displaystyle \frac{d\theta}{d\xi}=\omega \\[12pt]
\displaystyle \frac{d}{d\xi}\left(\xi^2 \omega\right)=-\xi^2 \, \theta^n
\quad \Rightarrow \quad \frac{d \omega}{d\xi} = -\theta^n - 2 \frac{\omega}{\xi}
\end{cases}
$$

$$
\frac{d}{d\xi} \left(\begin{matrix}
\theta \\
\omega
\end{matrix}\right) =
\left(\begin{matrix}
\omega \\
\displaystyle -\theta^n  - 2 \frac{\omega}{\xi}
\end{matrix}\right)
$$

$$
g_n := \frac{\rho}{\rho_c} = \theta^n(\xi_n) \, , \qquad 
f_n := -\xi_n^2\frac{d\theta}{d\xi}(\xi_n) 
$$

$$
h_n := 3 \, \frac{f_n}{\xi_n^3} \qquad 
\alpha_n := \frac{1}{n+1}\left(\frac{4\pi}{f_n^2}\right)^{1/3}
$$

```python
NOMI = np.array(["indice_n", "n", "csi_n", "theta_n", "theta_p_n", "g_n", "f_n", "h_n", "alpha_n"])
table = Table.read(f'build/Lane-Emden_zerosfunc.dat', names = NOMI,  format="ascii")
table
```

```python
ORDINE = {nome: i for i,nome in enumerate(NOMI)}
data = np.array( [ table[nome] for nome in NOMI  ] ).T
```

```python
ns = data[:, ORDINE["n"] ]
csins = data[:, ORDINE["csi_n"] ]
thetans = data[:, ORDINE["theta_n"] ]
thetans_p = data[:, ORDINE["theta_p_n"] ]
gns = data[:, ORDINE["g_n"] ]
fns = data[:, ORDINE["f_n"] ]
hns = data[:, ORDINE["h_n"] ]
alphans = data[:, ORDINE["alpha_n"] ]
```

```python
SIZE_ticks = 15
SIZE_labels = 20

fig, axs = plt.subplots(4, 2, figsize=(12, 24), tight_layout=True)
axs = axs.reshape(-1)

v_data = [csins, thetans, thetans_p, gns, fns, hns, alphans]
v_ynames = [r'$\xi_n$', r'$\theta(\xi_n)$', r'$\frac{d\theta}{d\xi}(\xi_n)$', 
            r'$g_n$', r'$f_n$', r'$h_n$', r'$\alpha_n$']

for i in range(len(v_data)):
    ax = axs[i]
    ax.plot(ns, v_data[i], '-', lw=3)
    ax.grid()
    #plt.xticks(np.arange(14,20,1), size=SIZE_2)
    #plt.yticks(np.arange(0,10,2), size=SIZE_2)
    ax.tick_params(axis='x', size=SIZE_ticks)
    ax.tick_params(axis='y', size=SIZE_ticks)
    ax.set_xlabel(r'$n$', size=SIZE_labels)
    ax.set_ylabel(v_ynames[i], size=SIZE_labels);
```
